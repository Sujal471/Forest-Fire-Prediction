# feature Selection

```python
def threshold(Dataframe,cutoff):
    set1=set()
    matrix=Dataframe.corr()
    for j in range(len(matrix.columns)):
        for i in range (j):
            if(matrix.iloc[i,j]>cutoff):
                set1.add(matrix.columns[j])
    return set1
        
    
```

```python
Remove=threshold(X_train,0.85)
print(Remove)
```

# # Standardization of data

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
X_train_scaled=scalar.fit_transform(X_train)
X_test_scaled=scalar.transform(X_test)
```

```python

plt.figure(figsize=(8,6))
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')

plt.show()
plt.figure(figsize=(8,6))
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
plt.show()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linmodel=LinearRegression()
```

```python
linmodel.fit(X_train_scaled,y_train)
y_pred=linmodel.predict(X_test_scaled)


```

```python
plt.plot(y_test,y_test,color='red',alpha=0.5)
plt.scatter(y_test,y_pred,color='blue',alpha=1)
plt.show()
```

```python
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("Mean absolute error : ",mae)
print("r2 score : ",r2)
```

# Lasso Regression


```python
from sklearn.linear_model import Lasso
Lassomodel=Lasso()
Lassomodel.fit(X_train_scaled,y_train)
y_pred=Lassomodel.predict(X_test_scaled)
plt.plot(y_test,y_test,color='red',alpha=0.5)
plt.scatter(y_test,y_pred,color='blue',alpha=1)
plt.show()
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("Mean absolute error : ",mae)
print("r2 score : ",r2)
```

```python
alphas=np.linspace(0.0001,5,200)
```

# Using Grid Search in Lasso Regression

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'alpha': alphas  # Testing 100 values between 0.0001 and 10
}
```

```python
lasso=Lasso()
grid_search = GridSearchCV(lasso, param_grid, scoring='neg_mean_squared_error', cv=5, n_jobs=-1)
grid_search.fit(X_train_scaled,y_train)
```

```python
print("Best Alpha:", grid_search.best_params_['alpha'])
print("Best MSE from CV:", -grid_search.best_score_)
```

```python
alphas = grid_search.cv_results_['param_alpha'].data  # Alpha values tested
mse_scores = -grid_search.cv_results_['mean_test_score']
print(len(alphas))
print(len(mse_scores))
```

```python
alphas
```

```python
coefficients=[]
for alpha in alphas:
    lasso=Lasso(alpha=alpha)
    lasso.fit(X_train_scaled,y_train)
    coefficients.append(lasso.coef_)
    
```

```python
coff=np.array(coefficients)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(alphas,mse_scores)
plt.xscale('log')
plt.show()
```

```python
plt.plot(alphas,coff)
plt.xscale('log')
plt.show()
```

```python
lasso= grid_search.best_estimator_
```

```python
lasso.alpha
```

```python
y_pred=lasso.predict(X_test_scaled)
```

```python
plt.plot(y_test,y_test,color='red',alpha=0.5)
plt.scatter(y_test,y_pred,color='blue',alpha=1)
plt.show()
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print("Mean absolute error : ",mae)
print("r2 score : ",r2)
```
